```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv(r'C:\Users\meetd\OneDrive\Desktop\Datasets\customer-transaction\train.csv')
test_data = pd.read_csv(r'C:\Users\meetd\OneDrive\Desktop\Datasets\customer-transaction\test.csv')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
print(train_data['target'].sum())
train_data.sort_values(by='target',ascending=False,inplace=True)
train_data.head()
train_data = train_data.iloc[:50000]
```

```python
print(train_data.shape)
print(test_data.shape)
train_data.drop(['ID_code'],axis=1,inplace=True)
test_data.drop(['ID_code'],axis=1,inplace=True)

x_test = test_data
x_train = train_data.loc[:, train_data.columns != 'target']
y_train = train_data.loc[:, train_data.columns == 'target']
print(x_train.shape, y_train.shape, x_test.shape)
```

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid")
missing = train_data.isnull().sum()
missing = missing[missing>0]
print(missing.values)
if(missing.any() == None):
    print('Missing features')
else:
    print('No missing features')
```

```python
dtype = set(train_data.dtypes)
print(dtype)
```

```python
y_train.nunique()
```

```python
import tensorflow as tf
```

```python
Model = tf.keras.Sequential([
    tf.keras.layers.Dense(1024,activation='relu'),
    tf.keras.layers.Dense(512,activation='relu'),
    tf.keras.layers.Dense(256,activation='relu'),
    tf.keras.layers.Dense(256,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
Model.compile(optimizer='adam',loss='binary_crossentropy', metrics='accuracy')
```

```python
Model.fit(x_train,y_train, batch_size=500, epochs=50,validation_split=0.2)
```

```python
Model.summary()
```

```python
y_pred = Model.predict(x_test)
print(y_pred)
```
